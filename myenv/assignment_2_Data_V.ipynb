```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,5,0.25)
y1 = x *2
y2 = x**2 * 2
y3 = 2**x * 5

plt.plot(x, y1, 'g--', x,y2, 'bs',x, y3, 'r^') # green dashed, blue squares and red triangle
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines with Different Styles")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
dates= pd.to_datetime(["03-10-16","04-10-16","05-10-16","06-10-16","07-10-16"],dayfirst=True)
closing_values= np.array([772.559998,776.429993,776.469971,776.859985,775.080017])

plt.plot(dates,closing_values,'r--o')
plt.xlabel("Date",fontsize=12,color='blue')
plt.ylabel("Closing Value",fontsize=12,color='blue')
plt.title("Closing Stock Value of Alphabet Inc.", fontsize=14,color='blue')

#customizing grid
plt.grid(which='major', linestyle='-', linewidth='0.5', color='blue')
plt.grid(which='minor', linestyle=':', linewidth='0.8', color='gray')

# highlighting a specific date
plt.axhline(y=773,color='blue', linestyle='-', linewidth=1.0)
plt.axhline(y=776,color='blue', linestyle='-', linewidth=1.0)

# Turn on minor ticks (for grid), but hide tick labels
plt.minorticks_off()
plt.tick_params(left=False, bottom=False, labelleft=True, labelbottom=True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%y-%m-%d'))
plt.xticks(rotation=45,fontsize=8)

plt.show()
```

```python
import matplotlib.pyplot as plt
sample_data = pd.DataFrame({
    'a': [2,4,8,5,7,6],   
    'b': [4,2,3,4,2,6],
    'c': [6,4,7,4,7,8],
    'd': [8,2,6,4,8,6],
    'e': [10,2,4,3,3,2]

})
colors = {'a': 'blue', 'b': 'green', 'c': 'red', 'd': 'lightgreen', 'e': 'purple'}
sample_data[['a', 'b', 'c','d','e']].plot(kind='bar', color=[colors[col] for col in ['a', 'b', 'c','d','e']])

plt.xlabel("indexes",color= "blue")
plt.ylabel("Values", color ="black")
plt.grid(True)
```

```python
# Import fresh (kernel restart recommended if plt still errors)
import pandas as pd
import matplotlib.pyplot as plt

# Create sample DataFrame
sample_data = pd.DataFrame({
    'a': [2, 4, 8, 5, 7, 6],
    'b': [4, 2, 3, 4, 2, 6],
    'c': [6, 4, 7, 4, 7, 8],
    'd': [8, 2, 6, 4, 8, 6],
    'e': [10, 2, 4, 3, 3, 2]
})

# Create grouped bar plot using axis-based method (more robust)
ax = sample_data[['a', 'b', 'c', 'd', 'e']].plot(kind='bar', figsize=(10, 6))
ax.set_xlabel('Indexes', color='blue')
ax.set_ylabel('Values', color='black')
ax.grid(True)
ax.set_title('Grouped Bar Plot of Sample Data')
plt.tight_layout()
plt.show()
```
