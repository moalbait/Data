```python
import numpy as np
import matplotlib as plt
x = np.arange(0,5,0.25)
y1 = x*2
y2 = x**2 * 2
y3 = 2**x * 5

plt.plot(x, y1, 'g--', y2, 'bs', y3, 'r^') # gree dashed, blue squares and red triangle
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines with Different Styles")
plt.grid(False)
plt.show()
```
