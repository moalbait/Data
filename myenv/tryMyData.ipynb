```python
%pip install openpyxl
import pandas as pd 
import numpy as np
from sqlalchemy import create_engine
from pathlib import Path
#update this to your folder path
base = Path(" ")

# =============================================================
# 2. IMPORTING DATA (CSV, Excel, JSON, Tab-delimited, Fixed-Width)
# =============================================================
df_csv = pd.read_csv("students.csv")
df_excel = pd.read_excel("students.xlsx", sheet_name="Sheet1")

df_json = pd.read_json("students.json")
df_tab = pd.read_csv("students_tab.txt", sep="\t")
df_fwf = pd.read_fwf("students_fwf.txt")
df = pd.concat([df_csv,df_excel,df_json,df_fwf,df_tab], ignore_index=True)

# print head of data 
print("\n--Head --")
print(df.head())
print("\n---INFO ---")
print(df.info())
print("\n--- DESCRIBE---")
print(df.describe())

# =============================================================
# 4. CLEANING THE DATA
# =============================================================

# Remove exact duplicate rows
df = df.drop_duplicates()
```
