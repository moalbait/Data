```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon
import pandas as pd

# 1. Create sample GeoDataFrame with Indonesian province-like geometry
provinces = {
    'NAME_1': ['Jakarta', 'Surabaya', 'Bandung', 'Medan', 'Makassar'],
    'population': [10562088, 2874087, 2481760, 2210624, 1614336]
}

# Sample geometries (rectangles representing provinces)
geometries = [
    Polygon([(106.65, -6.35), (107.15, -6.35), (107.15, -5.85), (106.65, -5.85)]),  # Jakarta
    Polygon([(112.5, -7.5), (113.0, -7.5), (113.0, -7.0), (112.5, -7.0)]),           # Surabaya
    Polygon([(107.0, -7.0), (107.5, -7.0), (107.5, -6.5), (107.0, -6.5)]),          # Bandung
    Polygon([(98.0, 3.5), (98.5, 3.5), (98.5, 4.0), (98.0, 4.0)]),                  # Medan
    Polygon([(119.0, -8.0), (119.5, -8.0), (119.5, -7.5), (119.0, -7.5)])           # Makassar
]

gdf = gpd.GeoDataFrame(provinces, geometry=geometries, crs='EPSG:4326')

# 2. Inspect the columns
print("Columns:", gdf.columns.tolist())
print("\nFirst few rows:")
print(gdf.head())

# 3. Plot the provinces
fig, ax = plt.subplots(figsize=(10, 8))
gdf.plot(ax=ax, edgecolor="black", facecolor="lightblue", alpha=0.7)

# 4. Add province labels using NAME_1
for idx, row in gdf.iterrows():
    plt.annotate(
        text=row["NAME_1"],
        xy=(row["geometry"].centroid.x, row["geometry"].centroid.y),
        ha="center", fontsize=9, fontweight='bold'
    )

plt.title("Indonesia Provinces (Level 1)", fontsize=14)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
